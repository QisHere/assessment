```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
```

```python
# import data and look at the first few rows

dfL1Data = pd.read_csv(r'./ABC_Tick_Data/ABC_Level_One_Tick_Data.csv')
dfL1Data.head()
```

```python
# convert Time_Hour to datetime objects and seperate date from time
dfL1Data['Time_Hour_Formatted'] = dfL1Data.Time_Hour.astype('datetime64')

dfL1Data['Date'] = dfL1Data.apply(lambda x: x['Time_Hour_Formatted'].date(),axis=1)
dfL1Data['Time'] = dfL1Data.apply(lambda x: x['Time_Hour_Formatted'].time(),axis=1)
```

```python
# check trading hours 
print("the earliest trade is made at {}".format(dfL1Data.Time.min()))
print("the latest trade is made at {}".format(dfL1Data.Time.max()))
```

```python
dfSpreadByHourStats = dfL1Data[['Avg_Bid_Ask_Spread','Time','Volume']].groupby(by='Time',as_index=True).describe()
dfSpreadByHourStats
```

* 8am-4pm seems to be the standard trading hour as trading outside these hours were not happening everyday
* trading after 5pm seems to happen very rarely but have the biggest spread on average and thus highest overall costs. Therefore trading data after 5pm will be removed from model training as they are outliers and may skew the output
* there are negative spreads in the data which should be rare and can arise either from extremely volatile or illiquid market (source: https://www.investopedia.com/terms/c/crossedmarket.asp#:~:text=A%20crossed%20market%20is%20the%20name%20traders%20and,with%20the%20advance%20of%20electronic%20and%20computerized%20trading)



```python
# explore negative spread in more details

dfNegSpread = dfL1Data[dfL1Data['Avg_Bid_Ask_Spread']<0]
dfNonNegSpread = dfL1Data[dfL1Data['Avg_Bid_Ask_Spread']>=0]
```

```python
dfNegSpread.describe()
```

```python
dfNonNegSpread.describe()
```

```python
print("{} shares traded at a negative spread".format(sum(dfNegSpread['Volume'])))
print("{} shares traded at a non negative spread".format(sum(dfNonNegSpread['Volume'])))
```

It seems from the data that trades with negative spread happened less frequently but they represented a little more than half of the trading volume and magnitude-wise, negative spreads are on average much wider than positive spread.

to account for the negative spread, one may try weigh the spread by trading volume then add them together. 

```python
# check feature correlation 1
dfCorr = dfL1Data.corr()
dfCorr
```

```python
# find the price that is most correlated to other prices

dfCorrPrice = dfCorr.loc[['Open','High','Low','Close','VWAP','TWAP'],['Open','High','Low','Close','VWAP','TWAP']]
dfCorrPrice['Total'] = dfCorrPrice.sum()
dfCorrPrice
```

```python
# check feature correlation 2
sns.pairplot(dfL1Data)
```

* Turnover and volume are highly positively correlated. This makes sense, the more shares traded, the more liquidity the asset has and the higher the turnover as a result. 

* Volume and spread are negatively correlated. This makes sense as more liquidity the asset has the narrower the spread will be

* All the prices are highly correlated with each other but none of them has a strong correlation with the spread, at least not linearly

* VWAP has the highest total correlation with all other price indicators.

* Volume data is highly skewed


From **data_exploration.ipynb**, the following statement was established:

1) Trading after 5pm seems to happen very rarely but have the biggest spread on average and thus highest overall costs. Therefore trading data after 5pm will be removed from model training as they are outliers and may skew the output

2) To account for the negative spread, one should weigh the positive and negative spread by trading volume then add them together

3) Turnover and volume are highly positively correlated

4) All the prices are highly correlated with each other but none of them has a strong correlation with the spread, at least not linearly

5) VWAP has the highest total correlation with all other price indicators.

6) Volume data is highly skewed

Therefore the following transformation should be applied:

1) remove trading data after 5pm

2) use weighted average spread as a proxy for transaction cost for a given period

3) Use only volume as a feature in modelling because it is more readily available when making predictions

4) Use VMAP as a feature and discard other price indicators

5) Log transform volume data
